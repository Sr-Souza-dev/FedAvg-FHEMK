## Set CKKS Params

```python
import sys
import os

sys.path.append(os.path.abspath('.')) 

from ckks.main import CKKS
import numpy as np

ckks = CKKS(
    N=4,
    int_precision=16,
    decimal_precision=32,
    sigma = 3,
    model_size= 4
)
```

### Generate and store keys

```python
sk = ckks.generate_keys() 
ckks.save_key(sk=sk)
sk = ckks.load_key()
```

### Try simple encryption and decryption

```python
v1 = np.array([2, 3, 1, 9])
v2 = np.array([4, 5, 7, 8])

c1 = ckks.encrypt(sk, v1)
c2 = ckks.encrypt(sk, v2)

c_sum = c1 + c2

print("Dec V1: ", ckks.decrypt(sk, c1))
print("Dec V2: ", ckks.decrypt(sk, c2))
print("Dec SUM: ", ckks.decrypt(sk, c_sum))
```

### Try phase encryption and decryption

```python
v1 = np.array([2, 3, 1])
v2 = np.array([4, 5, 7, 8])

c1 = ckks.encrypt_phase1(sk)
c1 = ckks.encrypt_phase2(c1, v1)

c2 = ckks.encrypt_phase1(sk)
c2 = ckks.encrypt_phase2(c2, v2)

c_sum = c1 + c2

print("Dec V1: ", ckks.decrypt(sk, c1))
print("Dec V2: ", ckks.decrypt(sk, c2))
print("Dec SUM: ", ckks.decrypt(sk, c_sum))
```

```python
x = 100
ckks = CKKS(
    N=1000,
    int_precision=16,
    decimal_precision=32,
    sigma = 3,
    model_size= x
)
```

### Try simple batch encryption and decryption

```python
v1 = np.array([2, 3, 1, 9, 2, 3, 1, 9, 2, 3, 1, 9, 2, 3, 1, 9, 3, 4, 5])
v2 = np.array([4, 5, 7, 8, 4, 5, 7, 8, 4, 5, 7, 8, 4, 5, 7, 8, 4, 5, 7])

c1 = ckks.encrypt_batch(sk, v1)
c2 = ckks.encrypt_batch(sk, v2)

c_sum = [ c_1 + c_2 for c_1, c_2 in zip(c1, c2)]

print("Dec V1: ", ckks.decrypt_batch(sk, c1))
print("Dec V2: ", ckks.decrypt_batch(sk, c2))
print("Dec SUM: ", ckks.decrypt_batch(sk, c_sum))
```

## Extracting coefficients to send

```python
v1 = np.array([2, 3, 1, 9, 2, 3, 1, 9, 2, 3, 1, 9, 2, 3, 1, 9, 3, 4, 5])
v2 = np.array([4, 5, 7, 8, 4, 5, 7, 8, 4, 5, 7, 8, 4, 5, 7, 8, 4, 5, 7])

c1 = ckks.encrypt_batch(sk, v1)
c1 = ckks.extract_vector(c1)            # extract coefficients to send throw the network
c1 = ckks.construct_cryptograms(c1)     # reconstruct the cryptogram from its coefficients

c2 = ckks.encrypt_batch(sk, v2)
c2 = ckks.extract_vector(c2)
c2 = ckks.construct_cryptograms(c2)

c_sum = [ c_1 + c_2 for c_1, c_2 in zip(c1, c2)]

print("Dec V1: ", ckks.decrypt_batch(sk, c1))
print("Dec V2: ", ckks.decrypt_batch(sk, c2))
print("Dec SUM: ", ckks.decrypt_batch(sk, c_sum))
```
