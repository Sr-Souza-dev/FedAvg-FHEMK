## DataLoader Test

```python
from fl_simulation.model.data_loader import load_data
import matplotlib.pyplot as plt
import numpy as np

train_data, test_data = load_data(1, 2)

import matplotlib.pyplot as plt
import torchvision
import torch

# Obter uma batch do seu DataLoader
batch = next(iter(train_data))
images = batch["image"]  # As imagens estão aqui
labels = batch["label"]

# Desnormalizar imagens (porque você usou Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)))
images = images * 0.5 + 0.5  # Volta para [0, 1]

# Função para mostrar as imagens
def imshow(imgs, labels):
    grid = torchvision.utils.make_grid(imgs[:8], nrow=8)
    npimg = grid.numpy().transpose((1, 2, 0))
    plt.figure(figsize=(12, 2))
    plt.imshow(npimg)
    plt.axis("off")
    plt.title("Labels: " + ", ".join(str(l) for l in labels[:8]))
    plt.show()

imshow(images, labels)
```

### Scheme Test

```python
from ckks.main import CKKS
import numpy as np

scheme = CKKS(
    N=4, 
    int_precision= 32, 
    decimal_precision=8, 
    sigma=3, 
    level=0
)

sk, pk, rlk = scheme.generate_keys()

v1 = np.array([2, 3])
v2 = np.array([4, 5])

c1 = scheme.encrypt_sk(sk, v1)
c2 = scheme.encrypt_sk(sk, v2)

sum_c = c1*5
sum_np = v1*5

dec_sum = scheme.decrypt(sk, sum_c)

print("Sum_np: ", sum_np)
print("Sum_c: ", dec_sum)
```
